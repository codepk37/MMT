```python
import os
import json
import csv

# Function to compile JSON files into CSV
def compile_json_to_csv(folder_path):
    # Iterate through files in the folder
    files = [f for f in os.listdir(folder_path) if f.endswith('.json')]
    files.sort()  # Sort files numerically
    csv_file_path = os.path.join(folder_path, 'compiled_data.csv')

    # Write data to CSV file
    with open(csv_file_path, 'w', newline='', encoding='utf-8') as csvfile:
        fieldnames = ['endTime', 'artistName', 'trackName', 'msPlayed']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()

        # Iterate through JSON files
        for file_name in files:
            json_file_path = os.path.join(folder_path, file_name)
            with open(json_file_path, 'r', encoding='utf-8') as jsonfile:
                data = json.load(jsonfile)
                for entry in data:
                    writer.writerow(entry)

    print(f"CSV file '{csv_file_path}' has been created successfully.")

# Main function to iterate through folders

# Iterate through folders
for folder_name in os.listdir():
    folder_path = os.path.join(os.getcwd(), folder_name)
    if os.path.isdir(folder_path) and folder_name.startswith('p'):
        if folder_name.endswith("110"):
            continue
        compile_json_to_csv(folder_path)

# 110
        
```

```python
# code for changing compiled_data.csv to compiled_data<folder_name>.csv
for folder_name in os.listdir():
    folder_path = os.path.join(os.getcwd(), folder_name)
    if os.path.isdir(folder_path) and folder_name.startswith('p'):
        if folder_name.endswith("110"):
            continue
        csv_file_path = os.path.join(folder_path, 'compiled_data.csv')
        new_csv_file_path = os.path.join(folder_path, f'compiled_data_{folder_name}.csv')
        os.rename(csv_file_path, new_csv_file_path)
        print(f"CSV file '{csv_file_path}' has been renamed to '{new_csv_file_path}' successfully.")


# copy all compiled_data<folder_name>.csv into a new file called compiled_data_all.csv
```
